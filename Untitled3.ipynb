```python
# Read the data from CSV file
df = pd.read_csv('/content/drive/My Drive/data_for_7034/portfolio_return_series.csv')

# Set risk-free rate (0.3% per month)
rf = 0.3

# Extract market returns from the dataframe
market_returns = df['MKT']
```

```python
# Calculate mean return and variance for market portfolio
mean_market = np.mean(market_returns)
var_market = np.var(market_returns)

# Current allocation (80% market, 20% risk-free)
w_market = 0.8
w_rf = 0.2

# Calculate portfolio mean and variance with current allocation
portfolio_mean = w_market * mean_market + w_rf * rf
portfolio_var = w_market**2 * var_market
```

```python
# Find risk aversion parameter A using the first-order condition
# First-order condition: mean_market - rf = A * w_market * var_market
A = (mean_market - rf) / (w_market * var_market)
```

```python
# Print initial statistics
print("Initial Portfolio Statistics:")
print(f"Market Mean Monthly Return: {mean_market:.4f}%")
print(f"Market Monthly Variance: {var_market:.4f}")
print(f"Risk-Free Monthly Rate: {rf:.4f}%")
print(f"Risk Aversion Parameter (A): {A:.6f}")
```

```python
# Generate weights array for plotting (0 to 1)
weights = np.linspace(0, 1, 101)

# Calculate certainty equivalent returns for different weights
ce_returns = []
for w in weights:
    mean_p = w * mean_market + (1-w) * rf
    var_p = w**2 * var_market
    ce = mean_p - 0.5 * A * var_p
    ce_returns.append(ce)

# Create single figure with all elements
plt.figure(figsize=(12, 8))

# Plot the CE returns curve
plt.plot(weights * 100, ce_returns, 'b-', label='Certainty Equivalent Returns', linewidth=2)
plt.axvline(x=80, color='r', linestyle='--', label='Current Allocation (80%)')

# Add labels and title
plt.xlabel('Weight in Market Index (%)')
plt.ylabel('Certainty Equivalent Return (%)')
plt.title('Monthly Certainty Equivalent Returns vs Market Weight')
plt.legend()
plt.grid(True)

# Add text box with key statistics in the same figure
stats_text = (f'Market Mean Return: {mean_market:.2f}%\n'
             f'Market Variance: {var_market:.2f}\n'
             f'Risk Aversion (A): {A:.4f}\n'
             f'Risk-Free Rate: {rf:.2f}%')
plt.text(0.02, 0.98, stats_text,
         transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', alpha=0.8),
         verticalalignment='top')

# Show the complete figure
plt.show()
```

```python
# Calculate and print optimal portfolio results
max_ce_idx = np.argmax(ce_returns)
max_ce = ce_returns[max_ce_idx]
optimal_weight = weights[max_ce_idx]

print("\nOptimal Portfolio Results:")
print(f"Maximum CE Return: {max_ce:.4f}%")
print(f"Optimal Weight in Market: {optimal_weight:.4f} ({optimal_weight*100:.1f}%)")

# Verify current portfolio efficiency
current_ce = portfolio_mean - 0.5 * A * portfolio_var
print(f"\nCurrent Portfolio (80/20) CE Return: {current_ce:.4f}%")
print(f"Difference from Optimal: {abs(current_ce - max_ce):.6f}%")
```

```python
def calculate_portfolio_metrics(weights, returns_df):
    """Calculate portfolio return and volatility"""
    portfolio_return = np.sum(returns_df.mean() * weights)
    portfolio_vol = np.sqrt(np.dot(weights.T, np.dot(returns_df.cov(), weights)))
    return portfolio_return, portfolio_vol
```

```python
def generate_random_portfolios(n_portfolios, assets_returns, long_only=True, weight_bounds=None):
    """Generate random portfolios based on constraints"""
    n_assets = len(assets_returns.columns)

    all_weights = []
    all_returns = []
    all_volatilities = []

    for _ in range(n_portfolios):
        if long_only:
            weights = np.random.random(n_assets)
            weights = weights / np.sum(weights)
        else:
            weights = np.random.uniform(weight_bounds[0], weight_bounds[1], n_assets)
            weights = weights / np.sum(np.abs(weights))  # Normalize to make sum of absolute weights = 1

        returns, vol = calculate_portfolio_metrics(weights, assets_returns)
        all_weights.append(weights)
        all_returns.append(returns)
        all_volatilities.append(vol)

    return np.array(all_weights), np.array(all_returns), np.array(all_volatilities)
```

```python
def plot_efficient_frontier(scenario_num, assets_returns, asset_names, long_only=True, weight_bounds=None):
    """Plot efficient frontier for a given scenario"""
    # Generate random portfolios
    weights, returns, volatilities = generate_random_portfolios(
        10000, assets_returns, long_only, weight_bounds)

    # Calculate individual asset metrics
    asset_returns = assets_returns.mean()  # This is a pandas Series
    asset_vols = pd.Series(
        np.sqrt(np.diag(assets_returns.cov())),
        index=asset_returns.index
    )  # Convert to pandas Series with same index

    # Calculate the tangent portfolio (maximum Sharpe ratio)
    sharpe_ratios = (returns - rf) / volatilities
    max_sharpe_idx = np.argmax(sharpe_ratios)
    tangent_weights = weights[max_sharpe_idx]

    # Plot
    plt.figure(figsize=(12, 8))
    plt.scatter(volatilities, returns, c='b', alpha=0.1, label='Portfolios')

    # Plot individual assets
    for asset_name in asset_names:
        plt.scatter(asset_vols[asset_name], asset_returns[asset_name],
                   c='r', marker='*', s=200,
                   label='Individual Assets' if asset_name==asset_names[0] else "")
        # Add asset labels
        plt.annotate(asset_name,
                    (asset_vols[asset_name], asset_returns[asset_name]),
                    xytext=(10, 10),
                    textcoords='offset points')

    # Plot tangent portfolio
    plt.scatter(volatilities[max_sharpe_idx], returns[max_sharpe_idx],
               c='g', marker='D', s=200, label='Tangent Portfolio')

    # Plot CAL
    x_range = np.linspace(0, max(volatilities)*1.2, 100)
    y_range = rf + (returns[max_sharpe_idx] - rf) * (x_range / volatilities[max_sharpe_idx])
    plt.plot(x_range, y_range, 'g--', label='Capital Allocation Line')

    plt.xlabel('Monthly Volatility (%)')
    plt.ylabel('Monthly Expected Return (%)')
    plt.title(f'Scenario {scenario_num}: Efficient Frontier and CAL')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print tangent portfolio weights
    print(f"\nScenario {scenario_num} - Tangent Portfolio Weights:")
    for asset, weight in zip(asset_names, tangent_weights):
        print(f"{asset}: {weight:.4f}")
    print(f"Tangent Portfolio Sharpe Ratio: {sharpe_ratios[max_sharpe_idx]:.4f}")
```

```python
import pandas as pd
import numpy as np
from datetime import datetime

def construct_factors():
    """
    Construct 5 new fundamental-based factors:
    1. Operating Efficiency = Operating Income (oiadpq) / Total Assets (atq)
        - Measures how efficiently a company generates operating profits from its assets
        - High efficiency indicates strong operational management

    2. Capital Investment = capxy / Total Assets (atq)
        - Measures investment intensity
        - Reflects company's growth potential and expansion strategy

    3. Working Capital Management = (actq - lctq) / saleq
        - Measures efficiency in managing short-term assets and liabilities
        - Reflects operational efficiency and liquidity management

    4. Asset Utilization = saleq / ppentq
        - Measures how efficiently fixed assets generate sales
        - Indicates asset productivity

    5. Financial Flexibility = cheq / lctq
        - Measures ability to meet short-term obligations
        - Indicates short-term solvency and financial safety
    """

    print(f"Starting at: {datetime.now()}")

    # 1. Load returns data for PS1 stock universe
    print("Loading returns data...")
    returns = pd.read_csv("/content/drive/My Drive/data_for_7034/monthly_stock_returns.csv",
                         dtype={'PERMNO': 'int32', 'MthRet': 'float32'})

    # Filter time period as in PS1
    returns = returns[
        (returns['YYYYMM'] >= 200001) &
        (returns['YYYYMM'] <= 202212)
    ].copy()

    valid_permnos = returns['PERMNO'].unique()

    # 2. Load quarterly fundamental data
    print("Loading fundamental data...")
    cols_needed = [
        'LPERMNO', 'datadate',
        'oiadpq',  # Operating Income
        'atq',     # Total Assets
        'actq',    # Current Assets
        'lctq',    # Current Liabilities
        'saleq',   # Sales
        'ppentq',  # Net Property, Plant and Equipment
        'cheq',    # Cash and Short-Term Investments
        'capxy'    # Capital Expenditure
    ]

    dtypes = {col: 'float32' for col in cols_needed if col not in ['LPERMNO', 'datadate']}
    dtypes['LPERMNO'] = 'int32'

    quarterly = pd.read_csv("/content/drive/My Drive/data_for_7034/fundamental_quarterly.csv",
                           usecols=cols_needed,
                           dtype=dtypes)

    # Filter for PS1 stock universe
    quarterly = quarterly[quarterly['LPERMNO'].isin(valid_permnos)].copy()
    quarterly['datadate'] = pd.to_datetime(quarterly['datadate'])

    # 3. Calculate fundamental factors
    print("Computing factors...")

    # Operating Efficiency: how well assets generate operating income
    quarterly['op_efficiency'] = quarterly['oiadpq'] / quarterly['atq']

    # Capital Investment: investment intensity relative to asset base
    quarterly['cap_investment'] = quarterly['capxy'] / quarterly['atq']

    # Working Capital Management: operational liquidity relative to sales
    quarterly['wc_management'] = (quarterly['actq'] - quarterly['lctq']) / quarterly['saleq']

    # Asset Utilization: sales generation efficiency from fixed assets
    quarterly['asset_util'] = quarterly['saleq'] / quarterly['ppentq']

    # Financial Flexibility: immediate liquidity coverage
    quarterly['fin_flexibility'] = quarterly['cheq'] / quarterly['lctq']

    # 4. Calculate monthly factor returns
    print("Computing monthly factor returns...")
    returns['date'] = pd.to_datetime(returns['YYYYMM'].astype(str), format='%Y%m')
    dates = sorted(returns['date'].unique())

    factors = ['op_efficiency', 'cap_investment', 'wc_management',
               'asset_util', 'fin_flexibility']

    factor_returns = pd.DataFrame(index=dates, columns=factors, dtype='float32')

    total_months = len(dates)
    for i, date in enumerate(dates):
        if i % 20 == 0:
            print(f"Processing month {i+1}/{total_months}")

        # Following the requirement: find most updated fundamental measurement at month t
        latest_data = quarterly[quarterly['datadate'] <= date].groupby('LPERMNO').last()
        month_rets = returns[returns['date'] == date].set_index('PERMNO')['MthRet']

        for factor in factors:
            combined = pd.DataFrame({
                'factor': latest_data[factor],
                'return': month_rets
            }).dropna()

            if len(combined) > 0:
                # Calculate top20% minus bottom20% equal-weighted portfolio returns
                quintiles = combined['factor'].quantile([0.2, 0.8])
                high_ret = combined[combined['factor'] >= quintiles[0.8]]['return'].mean()
                low_ret = combined[combined['factor'] <= quintiles[0.2]]['return'].mean()
                factor_returns.loc[date, factor] = high_ret - low_ret

    return factor_returns

if __name__ == "__main__":
    try:
        factor_returns = construct_factors()
        print("\nFactor Returns Summary:")
        print(factor_returns.describe().round(4))
        factor_returns.to_csv('new_fundamental_factors.csv')
        print("\nResults saved to CSV")
    except Exception as e:
        print(f"Error occurred: {str(e)}")
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from datetime import datetime

def load_and_process_data():
    """
    Load and process all required data with consistent data types and date handling
    """
    # Load FF5 factors - fix the header issue
    ff_factors = pd.read_csv('/content/drive/My Drive/data_for_7034/F-F_Research_Data_5_Factors_2x3.csv',
                            skiprows=3)  # Changed from skiprows=4
    ff_factors.columns = ['Month', 'Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA', 'RF']
    ff_factors = ff_factors.iloc[1:]  # Skip the header row

    # Convert FF5 numeric columns
    for col in ff_factors.columns[1:]:  # Skip 'Month' column
        ff_factors[col] = pd.to_numeric(ff_factors[col], errors='coerce') / 100

    # Load Q5 factors
    q5_factors = pd.read_csv('/content/drive/My Drive/data_for_7034/q5_factors_monthly_2023.csv')
    q5_factors['Month'] = q5_factors['year'].astype(str) + q5_factors['month'].astype(str).str.zfill(2)

    # Convert Q5 numeric columns
    numeric_cols = q5_factors.columns.difference(['Month', 'year', 'month'])
    q5_factors[numeric_cols] = q5_factors[numeric_cols].astype(float) / 100

    # Load liquidity factors
    liq_factors = pd.read_csv('/content/drive/My Drive/data_for_7034/liq_data_1962_2023.txt',
                             sep='\s+', comment='%', header=None)
    liq_factors.columns = ['Month', 'Agg_Liq', 'Innov_Liq_eq8', 'Traded_Liq_LIQ_V']

    # Convert liquidity numeric columns
    liq_factors[['Agg_Liq', 'Innov_Liq_eq8', 'Traded_Liq_LIQ_V']] = \
        liq_factors[['Agg_Liq', 'Innov_Liq_eq8', 'Traded_Liq_LIQ_V']].astype(float)

    # Load returns data
    returns = pd.read_csv("/content/drive/My Drive/data_for_7034/monthly_stock_returns.csv")
    returns.rename(columns={'YYYYMM': 'Month'}, inplace=True)
    returns['MthRet'] = returns['MthRet'].astype(float)

    # Filter time period
    start_date, end_date = '200001', '202212'
    date_mask = lambda x: (x['Month'].astype(str) >= start_date) & (x['Month'].astype(str) <= end_date)

    ff_factors = ff_factors[date_mask(ff_factors)]
    q5_factors = q5_factors[date_mask(q5_factors)]
    liq_factors = liq_factors[date_mask(liq_factors)]
    returns = returns[date_mask(returns)]

    # Create date column
    for df in [ff_factors, q5_factors, liq_factors, returns]:
        df['date'] = pd.to_datetime(df['Month'], format='%Y%m')

    return ff_factors, q5_factors, liq_factors, returns

def fama_macbeth_analysis():
    """
    Perform Fama-MacBeth analysis with both PS1 factors and new fundamental factors
    """
    print("Loading data...")
    ff_factors, q5_factors, liq_factors, returns = load_and_process_data()

    # Merge PS1 factors
    factors = ff_factors.merge(q5_factors[['date', 'R_ME', 'R_IA', 'R_ROE', 'R_EG']], on='date')
    factors = factors.merge(liq_factors[['date', 'Traded_Liq_LIQ_V']], on='date')

    # Add new fundamental factors
    new_factors = pd.read_csv('new_fundamental_factors.csv', index_col=0)
    new_factors.index = pd.to_datetime(new_factors.index)
    factors = factors.merge(new_factors, left_on='date', right_index=True)

    # Calculate excess returns
    returns['ExcessReturn'] = returns['MthRet'] - factors.set_index('date').loc[returns['date'], 'RF'].values

    # Define factor groups
    ps1_factors = ['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA',
                   'R_ME', 'R_IA', 'R_ROE', 'R_EG',
                   'Traded_Liq_LIQ_V']

    new_factors = ['op_efficiency', 'cap_investment', 'wc_management',
                   'asset_util', 'fin_flexibility']

    # Method 1: Combined analysis
    print("\nMethod 1: Combined Analysis")
    all_factors = ps1_factors + new_factors
    combined_results = run_fama_macbeth(returns, factors, all_factors)

    # Method 2: Separate analyses
    print("\nMethod 2: Separate Analyses")
    ps1_results = run_fama_macbeth(returns, factors, ps1_factors)
    new_factor_results = run_fama_macbeth(returns, factors, new_factors)

    return combined_results, ps1_results, new_factor_results

def run_fama_macbeth(returns, factors, factor_cols):
    """
    Helper function to run Fama-MacBeth regression for a given set of factors
    """
    print(f"Running analysis for {len(factor_cols)} factors...")
    stocks = returns['PERMNO'].unique()
    window = 60

    # First pass regression
    betas_list = []
    for stock in stocks:
        stock_data = returns[returns['PERMNO'] == stock]
        if len(stock_data) < window:
            continue

        stock_rets = stock_data.set_index('date')['ExcessReturn']
        stock_betas = pd.DataFrame(index=stock_data['date'])

        for factor in factor_cols:
            factor_data = factors.set_index('date')[factor]
            common_idx = stock_rets.index.intersection(factor_data.index)

            if len(common_idx) < window:
                continue

            roll = pd.DataFrame({
                'ret': stock_rets[common_idx],
                'factor': factor_data[common_idx]
            })

            cov = roll.rolling(window).cov().unstack()
            var = cov['factor']['factor']
            var = var.replace(0, np.nan)
            beta = cov['ret']['factor'] / var

            stock_betas[factor] = beta

        if not stock_betas.empty:
            stock_betas['PERMNO'] = stock
            betas_list.append(stock_betas)

    betas = pd.concat(betas_list)

    # Second pass regression
    dates = sorted(returns['date'].unique())
    gammas = pd.DataFrame(index=dates, columns=factor_cols, dtype=float)
    r2s = pd.Series(index=dates, dtype=float)

    for date in dates:
        date_returns = returns[returns['date'] == date]
        date_betas = betas.loc[date].dropna()

        common_stocks = np.intersect1d(date_returns['PERMNO'], date_betas['PERMNO'])
        if len(common_stocks) < 10:
            continue

        y = date_returns[date_returns['PERMNO'].isin(common_stocks)].set_index('PERMNO')['ExcessReturn']
        X = date_betas[date_betas['PERMNO'].isin(common_stocks)].set_index('PERMNO')[factor_cols]

        if len(y) > 0 and not X.isnull().any().any():
            model = sm.OLS(y, X).fit()
            gammas.loc[date] = model.params
            r2s.loc[date] = model.rsquared

    # Calculate statistics
    gammas = gammas.dropna(how='all')
    r2s = r2s.dropna()

    results = pd.DataFrame({
        'Mean Premium': gammas.mean(),
        't-statistic': gammas.mean() / (gammas.std() / np.sqrt(len(gammas))),
    })
    results['p-value'] = 2 * (1 - stats.t.cdf(abs(results['t-statistic']), len(gammas)-1))
    results['R-squared'] = r2s.mean()

    print(f"\nResults for {len(factor_cols)} factors:")
    print(results)
    print(f"Average R-squared: {r2s.mean():.4f}")

    return results
```

```python
def create_visualizations(combined_results, ps1_results, new_results):
    """
    Create and save visualization plots for the analysis results
    """
    print("\nCreating visualizations...")

    # Use default style
    plt.style.use('default')

    # 1. R-squared Comparison
    plt.figure(figsize=(10, 6))
    r2_values = {
        'Combined Model': combined_results['R-squared'].iloc[0],
        'PS1 Factors': ps1_results['R-squared'].iloc[0],
        'New Factors': new_results['R-squared'].iloc[0]
    }
    plt.bar(r2_values.keys(), r2_values.values(), color=['#2ecc71', '#3498db', '#e74c3c'])
    plt.title('Model Comparison: R-squared Values', fontsize=12)
    plt.ylabel('R-squared')
    plt.ylim(0, 1)
    for i, v in enumerate(r2_values.values()):
        plt.text(i, v + 0.01, f'{v:.4f}', ha='center')
    plt.savefig('r2_comparison.png')
    plt.close()

    # 2. Factor Significance Comparison
    plt.figure(figsize=(15, 8))
    all_factors = pd.DataFrame({
        't-statistic': combined_results['t-statistic'],
        'p-value': combined_results['p-value']
    })

    plt.scatter(range(len(all_factors)), all_factors['t-statistic'],
               c=all_factors['p-value'], cmap='RdYlGn_r',
               s=100, alpha=0.6)
    plt.axhline(y=1.96, color='g', linestyle='--', alpha=0.5, label='5% Significance')
    plt.axhline(y=-1.96, color='g', linestyle='--', alpha=0.5)
    plt.axhline(y=0, color='gray', linestyle='-', alpha=0.3)

    plt.xticks(range(len(all_factors)), all_factors.index, rotation=45, ha='right')
    plt.title('Factor Significance Analysis', fontsize=12)
    plt.ylabel('t-statistic')
    plt.xlabel('Factors')
    plt.colorbar(label='p-value')
    plt.tight_layout()
    plt.savefig('factor_significance.png')
    plt.close()

    # 3. Mean Premium Comparison
    plt.figure(figsize=(15, 8))
    mean_premiums = combined_results['Mean Premium']
    t_stats = combined_results['t-statistic']
    std_errors = mean_premiums / t_stats

    plt.errorbar(range(len(mean_premiums)), mean_premiums,
                yerr=1.96*std_errors, fmt='o', capsize=5,
                color='#3498db', alpha=0.6)

    plt.axhline(y=0, color='gray', linestyle='-', alpha=0.3)
    plt.xticks(range(len(mean_premiums)), mean_premiums.index, rotation=45, ha='right')
    plt.title('Factor Risk Premiums with 95% Confidence Intervals', fontsize=12)
    plt.ylabel('Mean Premium')
    plt.xlabel('Factors')
    plt.tight_layout()
    plt.savefig('factor_premiums.png')
    plt.close()

    print("\nResults and visualizations saved")
```

```python
if __name__ == "__main__":
    try:
        print("Starting Fama-MacBeth analysis...")
        combined_results, ps1_results, new_results = fama_macbeth_analysis()

        # Save numerical results
        with pd.ExcelWriter('fama_macbeth_comparison.xlsx') as writer:
            combined_results.to_excel(writer, sheet_name='Combined_Analysis')
            ps1_results.to_excel(writer, sheet_name='PS1_Factors')
            new_results.to_excel(writer, sheet_name='New_Factors')

        # Create visualizations
        create_visualizations(combined_results, ps1_results, new_results)

    except Exception as e:
        print(f"Error occurred: {str(e)}")
```
